Hipótese descrita: FLUXO E CONVERSÃO DAS ÓTICAS DE RUA 
Dados relacionados:   Bairro, estado, Id loja, soma de valor total da loja, tipo de loja
Hipótese: O tipo de loja exerce influência direta sobre o valor total de vendas.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../../database/dataset gerado/base_vendas_limpa.csv")
df.head()
```

```python
STORE_TYPE_COL = "Dim_Lojas.Tipo_PDV"    
SALES_COL = "Total_Preco_Liquido"       
```

```python
#agrupando e somando o valor de vendas
df_grouped = df.groupby(STORE_TYPE_COL)[SALES_COL].sum().reset_index()

#ordenando para visualização mais clara
df_grouped = df_grouped.sort_values(by=SALES_COL, ascending=False)
df_grouped
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df_grouped, x=STORE_TYPE_COL, y=SALES_COL, palette="viridis")

#ajustes visuais
plt.xticks(rotation=45, ha="right") #deixa a legenda inclinada
plt.title("Total de Vendas por Tipo de Loja", fontsize=14)
plt.xlabel("Tipo de Loja")
plt.ylabel("Total de Vendas")
plt.show()
```

Conclusão: Com base no gráfico gerado, é possível comprovar a seguinte hipótese: O tipo de loja exerce influência direta sobre o valor total de vendas. Essa hipótese diz sobre a influência direta que as categorias de loja têm sobre o valor líquido arrecadado por elas, porém, não é possível estimar qual exatamente é o tipo de loja que possuí o maior valor, uma vez que o gráfico apresenta incertezas, vista a variedade de lojas encontradas, algo incoerente, já que a empresa parceira não dispõe de tantas variedades. Isso constata uma incerteza nos nomes plotados.
