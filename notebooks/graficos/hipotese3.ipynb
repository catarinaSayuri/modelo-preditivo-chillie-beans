Hipótese descrita: CRESCIMENTO DA CATEGORIA DE GRAU
Dados relacionados: Tipo de loja, categoria, ticket medio de grau vs de sol (preço varejo)
Hipótese: A categoria de óculos de grau apresenta ticket médio superior à de óculos de sol, independentemente do tipo de loja.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../../database/dataset gerado/base_vendas_limpa.csv")
#filtra apenas as palvras-chave relevantes para a análise
filtro = df[df["Dim_Produtos.Grupo_Produto"].str.contains("grau|sol", case=False, na=False)]
df.head()
```

```python
#verifica o nome exato das colunas
df.columns
```

```python
CATEGORY_COL = "Dim_Produtos.Grupo_Produto"      
SALES_COL = "Preco_Varejo"     

#ticket médio por categoria
ticket_medio_categoria = df.groupby(CATEGORY_COL)[SALES_COL].mean().reset_index()

ticket_medio_categoria
```

```python
#vê qual categoria tem maior ticket médio
ticket_medio_categoria.sort_values(SALES_COL, ascending=False)
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x=CATEGORY_COL, y=SALES_COL, data=ticket_medio_categoria, palette="mako")
plt.title("Ticket Médio por Categoria (Grau vs Sol)", fontsize=14)
plt.xticks(rotation=90, ha="right")  
plt.xticks(rotation=90, ha="center") 
plt.xlabel("Categoria")
plt.ylabel("Ticket Médio (R$)")
plt.show()
```

Conclusão: Com base na observação e análise do gráfico, não é possível tirar conclusões muito sólidas para comprovar ou não a seguinte hipótese: A categoria de óculos de grau apresenta ticket médio superior à de óculos de sol, independentemente do tipo de loja; uma vez que, mesmo com a filtragem por palavras-chave para identificar óculos de sol e óculos de grau, ainda sim a base de dados entrega outros produtos ou palavras desconexas com o pedido da plotagem do gráfico.
